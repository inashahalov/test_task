```python
import pandas as pd

# Вариант 1: Без префикса data/
for year in [2020, 2021, 2022, 2023]:
    try:
        df = pd.read_csv(f'raw_{year}.csv', sep=';', nrows=1)
        has_deaths = 'Deaths' in df.columns
        print(f"{year}: {df.shape[1]} колонок | Deaths: {'✅' if has_deaths else '❌'}")
    except FileNotFoundError:
        print(f"{year}: ❌ Не найден")
```
