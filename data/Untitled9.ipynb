```python
import pandas as pd

df = pd.read_csv('cleaned_2020_with_daily.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Агрегируем по СТРАНАМ (суммируем все провинции)
country_daily = df.groupby(['Country/Region', 'Date'])['New_Cases'].sum().reset_index()

print("=" * 50)
print("📈 ТОП-5 СТРАН ПО ПИКУ ЗА ДЕНЬ (агрегировано)")
print("=" * 50)
top_countries = country_daily.groupby('Country/Region')['New_Cases'].max().sort_values(ascending=False).head(10)
print(top_countries)

print("\n" + "=" * 50)
print("🇨🇳 КИТАЙ — ПИК ПО ДНЯМ (все провинции вместе)")
print("=" * 50)
china = country_daily[country_daily['Country/Region'] == 'China']
china_peak = china.groupby('Date')['New_Cases'].sum().sort_values(ascending=False).head(10)
print(china_peak)
```
