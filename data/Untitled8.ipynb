```python
import pandas as pd

df = pd.read_csv('cleaned_2020_with_daily.csv')
df['Date'] = pd.to_datetime(df['Date'])

print("=" * 50)
print("1. СКОЛЬКО СТРАН В ДАННЫХ?")
print("=" * 50)
print(f"Уникальных стран: {df['Country/Region'].nunique()}")
print(df['Country/Region'].value_counts().head(10))

print("\n" + "=" * 50)
print("2. КИТАЙ — ПИК ЗА ДЕНЬ (по дням)")
print("=" * 50)
china = df[df['Country/Region'] == 'China']
china_peak = china.groupby('Date')['New_Cases'].sum().sort_values(ascending=False).head(10)
print(china_peak)

print("\n" + "=" * 50)
print("3. ВСЕГО ЗАПИСЕЙ ПО СТРАНАМ")
print("=" * 50)
country_counts = df.groupby('Country/Region').size().sort_values(ascending=False).head(15)
print(country_counts)
```
