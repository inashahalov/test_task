```python
import pandas as pd
import warnings

# Подавляем предупреждение о датах
warnings.filterwarnings('ignore')

# 1. Читаем и чистим сразу
df = pd.read_csv('raw_2020.csv', sep=';')
df = df.drop(columns=[c for c in df.columns if 'Unnamed' in c], errors='ignore')

# 2. Колонки с датами
date_cols = df.columns[2:]

# 3. Конвертируем в числа
for col in date_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

# 4. Melt (Wide → Long)
df_long = df.melt(
    id_vars=['Country/Region', 'Province/State'],
    value_vars=date_cols,
    var_name='Date',
    value_name='Confirmed'
)

# 5. Парсим даты
df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')

# 6. Фильтруем корректные данные
df_clean = df_long.dropna(subset=['Date', 'Confirmed']).copy()

# 7. Сохраняем результат (БЕЗ папки data/, мы уже внутри!)
df_clean.to_csv('cleaned_2020.csv', index=False)

print("=" * 50)
print("✅ ГОТОВО")
print("=" * 50)
print(f"Строк до очистки: {df_long.shape[0]}")
print(f"Строк после: {df_clean.shape[0]}")
print(f"Потери: {df_long.shape[0] - df_clean.shape[0]} строк")
print("\nДиапазон дат:")
print(f"  С: {df_clean['Date'].min()}")
print(f"  По: {df_clean['Date'].max()}")
print("\n📁 Файл сохранен: cleaned_2020.csv")
```
