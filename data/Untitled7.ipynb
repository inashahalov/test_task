```python
import pandas as pd

# Загружаем очищенные данные
df = pd.read_csv('cleaned_2020.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Считаем новые случаи за день (по каждой стране отдельно!)
df['New_Cases'] = df.groupby(['Country/Region', 'Province/State'])['Confirmed'].diff().fillna(0)

# Отрицательные значения = корректировки данных (оставляем как 0)
df['New_Cases'] = df['New_Cases'].apply(lambda x: max(0, x))

print("=" * 50)
print("📊 ПРИМЕР: Afghanistan (первые 10 дней)")
print("=" * 50)
afg = df[df['Country/Region'] == 'Afghanistan'][['Date', 'Confirmed', 'New_Cases']].head(10)
print(afg.to_string())

print("\n" + "=" * 50)
print("📈 ТОП-5 СТРАН ПО МАКСИМУМУ ЗА ДЕНЬ (2020)")
print("=" * 50)
top_daily = df.groupby('Country/Region')['New_Cases'].max().sort_values(ascending=False).head(5)
print(top_daily)

# Сохраняем финальную версию
df.to_csv('cleaned_2020_with_daily.csv', index=False)
print("\n✅ Файл сохранен: cleaned_2020_with_daily.csv")
```
