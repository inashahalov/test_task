```python
import pandas as pd

# Читаем
df = pd.read_csv('raw_2020.csv', sep=';')

# 1. Удаляем мусорную колонку
df = df.drop(columns=['Unnamed: 347'])

# 2. Проверяем первую дату на проблемы
print("Проблемные значения в первой дате:")
print(df['1/22/20'].value_counts().head(10))

# 3. Преобразуем все колонки с датами в числа (принудительно)
date_cols = df.columns[2:]  # всё после Country и Province
for col in date_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

# 4. Делаем melt (wide → long format)
df_long = df.melt(
    id_vars=['Country/Region', 'Province/State'],
    value_vars=date_cols,
    var_name='Date',
    value_name='Confirmed'
)

# 5. Чистим даты
df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')

print("\n" + "=" * 50)
print("ИТОГОВАЯ ТАБЛИЦА:")
print("=" * 50)
print(f"Строк: {df_long.shape[0]}")
print(df_long.head(10))
print("\nТипы данных:")
print(df_long.dtypes)
```
